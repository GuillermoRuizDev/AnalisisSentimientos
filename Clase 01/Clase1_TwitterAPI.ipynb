```python
from twython import Twython
```

```python
# Cargara las credenciales 
with open("twitter_credentials.json", "r") as file:
    creds = json.load(file)
```

```python
# Crear un Twython object
twython_obj = Twython(creds['CONSUMER_KEY'], creds['CONSUMER_SECRET'])
```

```python
# Query de bÃºsqueda
query = {'q': 'adele',
         #'geocode': ['-12.0580114','-77.0769462','1000km'],
         #'lang':'es',
         'result_type': 'recent',
         'count': 100
        }

search_results = twython_obj.search(**query)
```

```python
search_results.keys()
# statuses: informaciÃ³n recolectada
# search_metadata: parÃ¡metros de la bÃºsqueda
```

#### Convertir los textos a un dataset estructurado

```python
# Convertir la respuesta del API a un dataframe
tweets = []
for result in search_results['statuses']:
    tweet_dict = {}
    tweet_dict['id'] = result['id']
    tweet_dict['text'] = result['text']
    tweet_dict['truncated'] = result['truncated']
    tweet_dict['source'] = [re.findall('\>(.*?)\<', result['source'])[0] if result['source'] else ' ']
    tweet_dict['retweet_count'] = result['retweet_count']
    tweet_dict['favorite_count'] = result['favorite_count']
    tweets.append(tweet_dict)
df_tweets = pd.DataFrame(tweets)
df_tweets.shape
```
