### 1. Leer la información de la página de xkcd

```python
import urllib.request  as urllib2

url2='https://xkcd.com/archive/'
r = urllib2.urlopen(url2)
html = r.read()
print(html[:500])
```

### 2. Obtener información sobre cada tira cómica a partir del html utilizando regular expresions (regex)

```python
#https://docs.python.org/2/library/re.html

import re
htmltext = html.decode('utf-8')
strips = re.findall('\n<a href=(.*?)</a>', htmltext)

print ("len:     ",len(strips))
print ("example: ",strips[1])
```

### 3. Imprimir los 5 primeros resultados

```python
for i in strips[0:5]: print (i)
```

### 4. Imprimir  el número de tiras cómicas

```python
len(strips)
strips[100]
```

### 5. A partir de la información obtenida, crear una lista de nombre "info"

```python
info = []
for item in strips[0:2292]:
    n = re.findall('"/(.*?)/"',item)[0]
    date = re.findall('title=(.*?)>',item)[0]
    title = re.findall('>(.*?)$',item)[0]
    info.append([n,date,title])
print (info[100])
print (info[0:3])
```

### 6. Obtener automáticamente el url de la imagen 1027

```python
#código acá
info = []
for item in strips[0:2292]:
    n = re.findall('"/(.*?)/"',item)[0]
    date = re.findall('title=(.*?)>',item)[0]
    title = re.findall('>(.*?)$',item)[0]
    title = title.replace(":", "").replace("-", "_")
    url = 'https://xkcd.com/'+n
    splitTitle = title.lower().replace(' ','_')
    url2 = 'https://imgs.xkcd.com/comics/'+splitTitle+'.png'
    info.append([n,date,title,url,splitTitle,url2])
print (info[0:3])
print (len(info))
```

### 7. Obtener URLs que contengan la palabra "coronavirus".

```python
coronaURL=[]
for item in info[1:2292]:
  if 'coronavirus' in item[2].lower() :coronaURL.append(item[5])
print (len(coronaURL))
```
